```python
import pandas as pd
import numpy as np


num_customers = 1000
data = {
    'customer_id': range(1, num_customers + 1),
    'age': np.random.randint(18, 70, num_customers),
    'income': np.random.randint(30000, 150000, num_customers),
    'family_status': np.random.choice(['single', 'married_no_children', 'married_with_children', 'divorced', 'widowed'], num_customers, p=[0.3, 0.2, 0.3, 0.1, 0.1]),
    'risk_tolerance': np.random.choice(['low', 'medium', 'high'], num_customers, p=[0.4, 0.4, 0.2]),
    'existing_policies': np.random.choice(['none', 'health', 'life','auto','home','disabilty','health, home', 'home, life', 'disability, health'],
                                          num_customers, p=[0.6, 0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,]),
    'major_life_event': np.random.choice(['none', 'marriage', 'new_child', 'house_purchase', 'new_job', 'retirement', 'divorce'], num_customers, p=[0.5, 0.1, 0.1, 0.1, 0.05, 0.05, 0.1])
}

customer_df = pd.DataFrame(data)


customer_df.to_csv('customer_data.csv', index=False)

print("customer_data.csv generated successfully!")
```

```python
import pandas as pd

class LifeStageAnalyzer:
    def analyze(self, customer_data):
        """Analyzes customer data to determine their life stage."""
        life_stage = []
        for index, customer in customer_data.iterrows():
            age = customer['age']
            family_status = customer['family_status']
            major_life_event = customer['major_life_event']

            if age < 25 and family_status == 'single':
                stage = 'young_single'
            elif 25 <= age < 40 and family_status == 'single':
                stage = 'adult_single'
            elif 25 <= age < 55 and family_status in ['married_no_children', 'married_with_children']:
                 if 'new_child' in major_life_event or 'house_purchase' in major_life_event:
                    stage = 'established_family_major_event'
                 else:
                    stage = 'established_family'
            elif age >= 55:
                stage = 'empty_nester_or_retiree'
            else:
                stage = 'other' # Catch-all for combinations not explicitly covered

            life_stage.append(stage)

        customer_data['life_stage'] = life_stage
        return customer_data
```

```python
customer_df = pd.read_csv('customer_data.csv')
customer_df.head()
```

```python
if customer_df is not None:
  analyzer = LifeStageAnalyzer()
  customer_df = analyzer.analyze(customer_df)

  display(customer_df.head())
  customer_df.to_csv('customer_data_with_life_stages.csv', index=False)
```
